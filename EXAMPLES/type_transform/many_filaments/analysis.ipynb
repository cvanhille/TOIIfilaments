```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Lbox = 200
Abox = Lbox*Lbox
```

```python
f1 = pd.read_csv('r0.00/slice.out', header = 0, sep = '\s+')
f2 = pd.read_csv('r-0.00/slice.out', header = 0, sep = '\s+')
f3 = pd.read_csv('r0.01/slice.out', header = 0, sep = '\s+')
```

```python
plt.figure(figsize = (8,6))
plt.plot(f1['v_realtime'], f1['Temp'], color = 'slategray', label = 'Repulsive')
plt.plot(f2['v_realtime'], f2['Temp'], color = 'darkviolet', label = 'Attractive')
plt.plot(f3['v_realtime'], f3['Temp'], color = 'royalblue', label = 'Intermittent')
plt.axhline(1.0, ls = '--', lw = 1.0, color = 'black')
plt.xlabel(r'time $t$')
plt.ylabel(r'temperature $T$')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize = (8,6))
plt.plot(f1['v_realtime'], f1['PotEng'], color = 'slategray', label = 'Repulsive')
plt.plot(f2['v_realtime'], f2['PotEng'], color = 'darkviolet', label = 'Attractive')
plt.plot(f3['v_realtime'], f3['PotEng'], color = 'royalblue', label = 'Intermittent')
# plt.axhline(1.0, ls = '--', lw = 1.0, color = 'black')
plt.xlabel(r'time $t$')
plt.ylabel(r'potential energy $U$ ($k_BT$)')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize = (8,6))
plt.plot(f1['v_realtime'], f1['KinEng'], color = 'slategray', label = 'Repulsive')
plt.plot(f2['v_realtime'], f2['KinEng'], color = 'darkviolet', label = 'Attractive')
plt.plot(f3['v_realtime'], f3['KinEng'], color = 'royalblue', label = 'Intermittent')
plt.axhline(1.0, ls = '--', lw = 1.0, color = 'black')
plt.xlabel(r'time $t$')
plt.ylabel(r'kinetic energy $E_c$ ($k_BT$)')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize = (8,6))
plt.plot(f1['v_realtime'], f1['Press'], color = 'slategray', label = 'Repulsive')
plt.plot(f2['v_realtime'], f2['Press'], color = 'darkviolet', label = 'Attractive')
plt.plot(f3['v_realtime'], f3['Press'], color = 'royalblue', label = 'Intermittent')
plt.axhline(0.0, ls = '--', lw = 1.0, color = 'black')
plt.xlabel(r'time $t$')
plt.ylabel(r'pressure $P$ ($k_BT \sigma^{-3}$)')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize = (8,6))
plt.plot(f1['v_realtime'], f1['Atoms']/Abox, color = 'slategray', label = 'Repulsive')
plt.plot(f2['v_realtime'], f2['Atoms']/Abox, color = 'darkviolet', label = 'Attractive')
plt.plot(f3['v_realtime'], f3['Atoms']/Abox, color = 'royalblue', label = 'Intermittent')
# plt.axhline(0.0, ls = '--', lw = 1.0, color = 'black')
plt.xlabel(r'time $t$')
plt.ylabel(r'density $\rho$ ($\sigma^{-2}$)')
plt.legend()
plt.tight_layout()
plt.show()
```
